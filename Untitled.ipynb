```python
import sqlite3
# establishing a database connection
con = sqlite3.connect('centraldb.sqlite')
```

```python
cursor=con.cursor()
query = """ 
        select p.fieldID, u.representation from pipe_properties p left join units u on p.nativeUnitID=u.unitID 
		where p.pipeID=(select distinct min(pipeID) from pipe_properties)
		"""
cursor.execute(query)
records = cursor.fetchall()
records
```

```python
cursor=con.cursor()
query = """update work_units set unitID=3009 where parameterID=1001"""
cursor.execute(query)
records = cursor.fetchall()
con.commit()
```

```python
cursor=con.cursor()
query = """ update work_units set unitID=du.unitID
            from default_us_units as du
            where work_units.parameterID=du.parameterID
        """
cursor.execute(query)
records = cursor.fetchall()
con.commit()
```

```python
%gui qt5 
from PyQt5.QtWidgets import QApplication,QWidget,QScrollArea, QTableWidget, QVBoxLayout,QTableWidgetItem
import pandas as pd
app = QApplication([])
win = QWidget()
scroll = QScrollArea()
layout = QVBoxLayout()
table = QTableWidget()
scroll.setWidget(table)
layout.addWidget(table)
win.setLayout(layout)    

import pandas as pd
df = pd.read_sql('select * from fields',con)

#df = pd.DataFrame({"a" : [4 ,5, 6],"b" : [7, 8, 9],"c" : [10, 11, 12]},index = [1, 2, 3])
table.setColumnCount(len(df.columns))
table.setRowCount(len(df.index))
for i in range(len(df.index)):
    for j in range(len(df.columns)):
        table.setItem(i,j,QTableWidgetItem(str(df.iloc[i, j])))

win.show()
```
